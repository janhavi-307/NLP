# Assignment 10 - Write a better auto-complete algorithm using an N-gram model (similar models are used for translation, determining the author of a text, and speech recognition)

```python
import re
import nltk
from collections import defaultdict, Counter
import numpy as np

nltk.download('punkt')
```

```python
text = """
Machine learning is transforming the world.
Machine learning models help solve complex problems.
Artificial intelligence and machine learning are closely related.
Deep learning is a subset of machine learning.
"""
```

```python
nltk.download('punkt_tab')

def preprocess(text):

    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)

    tokens = nltk.word_tokenize(text)

    return tokens

tokens = preprocess(text)

print(tokens[:20])
```

```python
def build_ngrams(tokens, n):

    ngrams = []

    for i in range(len(tokens)-n+1):
        ngrams.append(tuple(tokens[i:i+n]))

    return ngrams

unigrams = build_ngrams(tokens,1)
bigrams = build_ngrams(tokens,2)
trigrams = build_ngrams(tokens,3)

unigram_counts = Counter(unigrams)
bigram_counts = Counter(bigrams)
trigram_counts = Counter(trigrams)
```

```python
vocab = set(tokens)
vocab_size = len(vocab)

print("Vocabulary size:", vocab_size)
```

```python
def bigram_prob(w1, w2):

    numerator = bigram_counts[(w1,w2)] + 1
    denominator = unigram_counts[(w1,)] + vocab_size

    return numerator / denominator
```

```python
def autocomplete(word, top_k=5):

    suggestions = {}

    for v in vocab:
        suggestions[v] = bigram_prob(word, v)

    sorted_words = sorted(suggestions.items(), key=lambda x: x[1], reverse=True)

    return sorted_words[:top_k]
```

```python
autocomplete("machine")
```

```python
def trigram_prob(w1,w2,w3):

    numerator = trigram_counts[(w1,w2,w3)] + 1
    denominator = bigram_counts[(w1,w2)] + vocab_size

    return numerator/denominator
```

```python
def autocomplete_trigram(w1,w2,top_k=5):

    suggestions = {}

    for v in vocab:
        suggestions[v] = trigram_prob(w1,w2,v)

    sorted_words = sorted(suggestions.items(), key=lambda x: x[1], reverse=True)

    return sorted_words[:top_k]
```

```python
autocomplete_trigram("machine","learning")
```
